```python
import numpy as np
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.0f' % x)
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv('../data/to_trasform.csv')
```

```python
df.head(2)
```

# Aggregation by average TCP retransmission

```python
retransmission_total = df['retrans_packets_dl_b'] + df['retrans_packets_ul_b']
```

```python
df['avg_retransmission'] = retransmission_total/2
```

```python
retransmission =df.groupby(['MSISDN'])["avg_retransmission"].sum().reset_index(name='avg_retrans')
```

```python
retransmission.sort_values(['avg_retrans'], ascending=False).head(10)
```

```python
retransmission.sort_values(['avg_retrans'], ascending=False).tail(10)
```

# Aggregation by average throughput

```python
throughput_total = df['throughput_avg_dl_kpbs'] + df['throughput_avg_ul_kpbs']
```

```python
df['avg_throughput'] = throughput_total/2
```

```python
throughput =df.groupby(['MSISDN'])["avg_throughput"].sum().reset_index(name='avg_throughput')
```

```python
throughput.sort_values(['avg_throughput'], ascending=False).head(10)
```

```python
throughput.sort_values(['avg_throughput'], ascending=False).tail(10)
```

# Aggregation by average RTT

```python
rtt_total = df['avg_rtt_dl'] + df['avg_rtt_ul']
```

```python
df['avg_rtt'] = rtt_total/2
```

```python
rtt =df.groupby(['MSISDN'])["avg_rtt"].sum().reset_index(name='avg_rtt')
```

```python
rtt.sort_values(['avg_rtt'], ascending=False).head(10)
```

```python
rtt.sort_values(['avg_rtt'], ascending=False).tail(10)
```

# Aggregation by average phones

```python
user_phone=df.groupby(['MSISDN','phone_name'])["phone_name"].count().reset_index(name='phone_count')
```

```python
user_phone.sort_values(['phone_count'], ascending=False).head(10)
```

```python
user_phone.sort_values(['phone_count'], ascending=False).tail(10)
```

```python
experience  = pd.merge(user_phone,rtt ,on='MSISDN', how='left')
experience  = pd.merge(experience, throughput, on='MSISDN', how='left')
experience  = pd.merge(experience, retransmission, on='MSISDN', how='left')
```
